```python
import re
def remove_mentions(text):
    return re.sub(r'@\w+', '', text)

# Apply the function to the 'comment' column
total_data['textDisplay'] = total_data['textDisplay'].apply(remove_mentions)
```

```python
import pandas as pd
from bs4 import BeautifulSoup
import html
import re

def clean_text(text):
    # Remove HTML comments

    text = re.sub(r'<!--(.*?)-->', '', text)

    # Remove or handle specific formatting tags (e.g., <i> and <a>)
    text = re.sub(r'<i>(.*?)</i>', r'\1', text)  # Remove <i> tags
    text = re.sub(r'<a(.*?)>(.*?)</a>', r'\2', text)  # Remove <a> tags and keep the link text

    # Decode common HTML entities
    text = html.unescape(text)

    # Handle <br> tags by converting them into newline characters
    if "<br>" in text:
        text = BeautifulSoup(text, 'html.parser').get_text(separator='\n')
    else:
        text = BeautifulSoup(text, 'html.parser').get_text()

    return text


# Apply the cleaning function to the 'comment' column
total_data['textDisplay'] = total_data['textDisplay'].apply(clean_text)

```

```python
import emoji

def replace_emojis_with_words(text):
    text = emoji.demojize(text)
    return(text)



total_data['textDisplay'] = total_data['textDisplay'].apply(replace_emojis_with_words)
```

## Tokenizing words from text to numeric

```python
import pandas as pd
import nltk
import gensim.downloader as gdownload
import numpy as np
import math
from nltk.corpus import words
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
nltk.download('words')
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
wordset = set(words.words())
stopwordset = set(stopwords.words('english'))
```

```python
lemmatizer = WordNetLemmatizer()
def tokenizer(sentence):
    sentence = sentence.split()
    # print(sentence)
    tokenized = []
    for word in sentence:
        word = word.lower()
        #print(word)
        if word in wordset:
            # print('yes')
            if word not in stopwordset:
                # print('lalala')
                word = lemmatizer.lemmatize(word)
                tokenized.append(word)
            else:
                continue
        else:
            continue
    return tokenized
```

```python
def sentence_embed(sentence,model):
    length = len(sentence)
    embedding = np.zeros(50)
    for word in sentence:
        try:
            embedding += model[word]
        except:
            embedding = embedding
    embedding = embedding/length
    return embedding
```

## Sentiment Analysis

```python
from nltk.sentiment import SentimentIntensityAnalyzer
```

```python
import nltk
nltk.download('vader_lexicon')
```

```python
sia = SentimentIntensityAnalyzer()
```

```python
sentiments = []
sia = SentimentIntensityAnalyzer()
for token in df['textDisplay']:
    combined_sentence = token
    #print(combined_sentence)
    # Initialize the sentiment analyzer


    # Perform sentiment analysis on the combined sentence
    sentiment = sia.polarity_scores(combined_sentence)
    #print(sentiment)
    # Extract the compound sentiment score
    compound_score = sentiment['compound']

    # Interpret the sentiment score based on your needs
    # For example, you might consider positive if compound score > 0.2, negative if < -0.2, and neutral otherwise
    sentiment_label = 'positive' if compound_score > 0.2 else 'negative' if compound_score < -0.2 else 'neutral'

    # Print or use the overall sentiment label as needed
    sentiments.append(sentiment_label)
```
